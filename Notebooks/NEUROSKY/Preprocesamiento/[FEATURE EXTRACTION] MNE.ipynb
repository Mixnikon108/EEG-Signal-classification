```python
import mne
import numpy as np
import plotly.express as px

# Supongamos que tienes un objeto 'raw' con tus datos EEG cargados

# Selecciona un canal de EEG para visualizar (reemplaza 'EEG 1' con el nombre de tu canal)
channel_name = 'FP1'

# Define los parÃ¡metros de filtrado
l_freq = 0.5  # Frecuencia de corte baja
h_freq = 50.0  # Frecuencia de corte alta

# Aplica el filtrado a la seÃ±al
raw_filtered = raw.copy().filter(l_freq=l_freq, h_freq=h_freq)

# Extrae la seÃ±al original sin filtrar del canal seleccionado
data_original = raw.copy().pick_channels([channel_name]).get_data()[0]

# Extrae la seÃ±al filtrada del canal seleccionado
data_filtered = raw_filtered.copy().pick_channels([channel_name]).get_data()[0]

# Configura el grÃ¡fico para la seÃ±al EEG original
fig_original = px.line(x=raw.times, y=data_original, labels={'x': 'Tiempo (s)', 'y': 'Amplitud'},
                       title='SeÃ±al EEG Original', line_shape='linear')

# Configura el grÃ¡fico para la seÃ±al EEG despuÃ©s del filtrado
fig_filtered = px.line(x=raw_filtered.times, y=data_filtered, labels={'x': 'Tiempo (s)', 'y': 'Amplitud'},
                       title='SeÃ±al EEG DespuÃ©s del Filtrado', line_shape='linear')

# Muestra los grÃ¡ficos
fig_original.show()
fig_filtered.show()
```

```python
import mne
import numpy as np
import plotly.express as px

# Supongamos que tienes un objeto 'raw' con tus datos EEG cargados

# Selecciona un canal de EEG para visualizar (reemplaza 'EEG 1' con el nombre de tu canal)
channel_name = 'FP1'

# Define los parÃ¡metros de filtrado
l_freq = 0.5  # Frecuencia de corte baja
h_freq = 50.0  # Frecuencia de corte alta

# Aplica el filtrado a la seÃ±al
raw_filtered = raw.copy().filter(l_freq=l_freq, h_freq=h_freq)

# Extrae la seÃ±al original sin filtrar del canal seleccionado
data_original = raw.copy().pick_channels([channel_name]).get_data()[0]

# Extrae la seÃ±al filtrada del canal seleccionado
data_filtered = raw_filtered.copy().pick_channels([channel_name]).get_data()[0]

# Calcula la transformada de Fourier para la seÃ±al original
fft_original = np.fft.fft(data_original)
freqs_original = np.fft.fftfreq(len(fft_original), 1 / raw.info['sfreq'])

# Calcula la transformada de Fourier para la seÃ±al filtrada
fft_filtered = np.fft.fft(data_filtered)
freqs_filtered = np.fft.fftfreq(len(fft_filtered), 1 / raw_filtered.info['sfreq'])

# Calcula el espectro de frecuencia en dB
psd_original = 10 * np.log10(np.abs(fft_original) ** 2)
psd_filtered = 10 * np.log10(np.abs(fft_filtered) ** 2)

# Filtra solo la mitad positiva del espectro
positive_mask = freqs_original >= 0
freqs_original = freqs_original[positive_mask]
psd_original = psd_original[positive_mask]
freqs_filtered = freqs_filtered[positive_mask]
psd_filtered = psd_filtered[positive_mask]

# Configura el grÃ¡fico para el espectro de frecuencia original en dB
fig_original = px.line(x=freqs_original, y=psd_original, labels={'x': 'Frecuencia (Hz)', 'y': 'Potencia (dB)'},
                       title='Espectro de Frecuencia (dB) de la SeÃ±al EEG Original')

# Configura el grÃ¡fico para el espectro de frecuencia filtrada en dB
fig_filtered = px.line(x=freqs_filtered, y=psd_filtered, labels={'x': 'Frecuencia (Hz)', 'y': 'Potencia (dB)'},
                       title='Espectro de Frecuencia (dB) de la SeÃ±al EEG DespuÃ©s del Filtrado')

# Muestra los grÃ¡ficos
fig_original.show()
fig_filtered.show()
```

```python
# EJECUTAR AL FINAL

# Definir las bandas de frecuencia de interÃ©s
freq_bands = [(8, 13), (13, 30), (30, 50)]  # Por ejemplo, alfa, beta, gamma

# Calcular la densidad espectral de potencia (PSD)
spectrum = raw.compute_psd(method="welch")
data, freqs = spectrum.get_data(return_freqs=True)

# Calcular la potencia media en cada banda de frecuencia
power_band = []
for fmin, fmax in freq_bands:
    power_band.append(np.mean(data[:, (freqs >= fmin) & (freqs < fmax)], axis=1))
```
